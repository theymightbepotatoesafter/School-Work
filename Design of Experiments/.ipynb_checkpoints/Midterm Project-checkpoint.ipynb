```python
#imports

import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import pymc3 as pm
```

```python
tData = pd.read_csv('./GlobalTemperatures.csv') #reading the .csv into a pandas dataframe
tempData = tData.fillna(value = -999) # filling all the nan with -999
```

```python
def chooseMonth(monthNumbers): #chooses all land average temp data from any number of months that are chosen
    #initializing lists
    tempMonth = []
    run_ave = []
    
    #for loop that runs through all the input months and picks out the data for each month
    for month in range(len(monthNumbers)):
        for i in range(tempData.dt.size):
            if int(tempData.dt[i][5:7]) == monthNumbers[month]:
                tempMonth.append(tempData.LandAverageTemperature[i])
    
    #calculates running average
    for l in range(len(tempMonth)):
        run_ave.append(np.sum(tempMonth[:l]) / l)
    
    #replaces -999 values (see cell above) with the corresponding running average value
    for temp in range(len(tempMonth)):
        if tempMonth[temp] == -999:
            tempMonth[temp] = run_ave[temp]
    
    return tempMonth
```

```python
def uncertainty(monthNumbers): #Picks out the uncertainty on land average temp of all months input
    #init lists
    tempUncert = []
    
    #for loop that gets the data
    for month in range(len(monthNumbers)):
        for i in range(tempData.dt.size):
            if int(tempData.dt[i][5:7]) == monthNumbers[month]:
                tempUncert.append(tempData.LandAverageTemperatureUncertainty[i])
    
    #returning as a masked array because uncertainty shouldn't be averaged
    return np.ma.masked_values(tempUncert, value = -999)
```

```python
def tickPicker(numTicks, offset = 50): #picking tick values for the graphs. Didn't really work so didn't really get used
    #list init
    labels = []
    ticks = []
    mod = np.round(tempData.dt.size / numTicks)
    
    #picks the tick values
    for i in range(numTicks):
        #print(tempData.dt.size % numTicks)
        labels.append(tempData.dt[(i * mod) + offset][:4])
        ticks.append(i * mod)
    
    #returns the labels and the tick locations
    return labels, np.array(ticks)
```

```python
def runningAve(data): #keeps a running average of tempuratures. 
    #list init
    run_ave = []
    
    #for loop to generate the average
    for i in range(len(data)):
        run_ave.append(np.sum(data[:i]) / i)
    
    #return the running average as an array
    return np.array(run_ave)
```

```python
#generating an array of the years for convinience
years = np.arange(1750,2016)
```

```python
months = [6] #choosing the months I want to look at
labels, ticks = tickPicker(7, 200) # making the labels and ticks
tempMonth = chooseMonth(months) #LAT
tempUncert = uncertainty(months) #LAT uncertainty
run_ave = runningAve(tempMonth) #running LAT
#plt.figure(figsize = (50,20))
#plt.xticks(ticks/12, labels); #didn't really work so I abandon the idea
plt.scatter(years[109:], tempMonth[109:]); #plotting the data
#plt.scatter(years[109:], run_ave[109:]); #plotting the running LAT
#plt.scatter(years[109:], tempUncert[109:] + 13); #plotting the LAT uncertainty
```

```python
june_data = tempMonth
```

```python
with pm.Model() as temp_model:
    
    #setting where the switch point might be
    switchpoint = pm.DiscreteUniform('switchpoint', lower = years.min(), upper = years.max(), testval = 1980)

    #priors for the early and late rates 
    early_rate = pm.Normal('early_rate', mu = 10, sigma = 2)
    late_rate = pm.Normal('late_rate', mu = 10, sigma = 2)
    
    #this switches the rates depending on what year you're at and where the switch point is
    rate = pm.math.switch(switchpoint >= years, early_rate, late_rate)

    #The distribution of tempuratures with the likelihood being input as the observed data
    tempurature = pm.Normal('tempurature', mu = rate, observed=june_data)
```

```python
with temp_model:
    trace = pm.sample(10000, chains = 5, tune = 1000)
```

```python
pm.traceplot(trace);
```

As you can see with the early and late rate path plots, the posterior has been pretty thoroughly sampled. The switchpoint could use a bit more time but for the most part the posterior has been sampled well. The early and late rate distribtion curves look very good considering that there are 5 chains with 10,000 steps each. It's very obviously a strong likelihood function, the data, that is the cause of this. I also tried using different priors for the early and late rates, exponential, and got the same looking curves centerd almost exactly where they are centered right now. This is pretty strong evidence that our priors aren't skewing the data. Now this isn't exactly the rates of temurature change but rather the average tempurature. However the rates of change are calculated but they aren't the most accurate. The markdown below explains it in more detail

```python
early_ave = trace['early_rate'].mean()
late_ave = trace['late_rate'].mean()
```

```python
early_slope = early_ave/(trace['switchpoint'].mean() - 1750)
late_slope = late_ave/(2015 - trace['switchpoint'].mean())
```
