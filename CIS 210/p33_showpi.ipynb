```python
from turtle import *
import math as m
import random
```

```python
def isInCircle(x, y, r):
    """
    (float, float, float) -> bool
    
    User inputs a point, (x,y), and a radius, r, and is returned if the point is in the circle of radius r
    
    >>> isInCircle(0.123409, 0.4444, 1)
    True
    """
    if (x**2 + y**2 <= r**2):
        return True
    else:
        return False
```

```python
def reportPi(numDarts, approxPi):
    """
    (int, float) -> string and float
    
    User inputs number of darts to be thrown and the approximate value of pi
    that was calculated from numDarts and is returned the error on approxPi
    
    >>> reportPi(1000, 3.141516)
    
    """
    error = approxPi / m.pi
    print('With %i itterations' % numDarts)
    print('my approximate value for pi is: %f' % approxPi)
    print('math lib pi value is: %f' % m.pi)
    print('This is a %.2f percent error' % error)
```

```python
def drawBoard():
    """
    (none) -> none
    
    Utility function to draw the board the simulation will be displayed on
    
    >>> drawBoard()
    
    """
    wn = Screen()
    wn.setworldcoordinates(-2, -2, 2, 2)

    speed(0); hideturtle()
    penup()

    goto(-1, 0)
    pendown()
    goto(1, 0)
    penup()
    goto(0, 1)
    pendown()
    goto(0, -1)
    penup()
    goto(0, -1)

    return None
```

```python
def showMontePi(numDarts):
    """
    (int) -> turtle image
    
    User inputs the number of darts to be thrown and is returned an image
    with dots located at the coordinates generated by the random function
    
    >>> showMontePi(1000)
    *turtle image*
    """
    drawBoard()
    # pen should stay up for drawing darts
 
    inCircle = 0

    # throw the darts and check whether
    # they landed on the dart board and
    # keep count of those that do
    for i in range(numDarts):
        x = random.random()
        y = random.random()

        # show the dart on the board
        if isInCircle(x, y, 1):
            inCircle += 1
            color('blue')
        else:
            color('red')

        goto(x, y)
        dot()

    # calculate approximate pi
    approxPi = inCircle/numDarts * 4

    #exitonclick()

    return approxPi

#revise to call new isInCircle function as in project 3-2
#add reportPi function as in project 3-2
#add drawBoard function
#add main function
```

```python
def main():
    """Runs all of the functions"""
    darts = 100
    approxPi = showMontePi(darts)
    reportPi(darts, approxPi)
    
```

```python
main()
```
