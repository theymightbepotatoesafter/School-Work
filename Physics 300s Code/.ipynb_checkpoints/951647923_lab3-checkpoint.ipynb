The calibration slide was focused in the microscope and the view was found to be 383 +/- 5 microns aligned on the left side (11:33) and 283 +/- 5 microns aligned from the top. Unclumped spheres were found in the focus of the microscope and the timelapse was started (11:39) and the tempurature was recorded to be 23.6 +/- 0.2 C (11:41).

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import csv
import pandas as pd
%matplotlib inline
```

```python
#------------3.6.2 Dispersion---------------
```

```python
#the data is read in
with open("lab3_data.csv", 'r') as f:
    data = list(csv.reader(f, delimiter=","))
delta = np.array(data[:])
DX = []
DY = []
r = len(delta[:,1])
for i in range(r-1):
    DX.append(float(delta[i+1,0]))
    DY.append(float(delta[i+1,1]))

#lists are created to hold the data so math can be done on them

Dx = np.array(DX)
Dy = np.array(DY)

dx = []
dy = []
dx1 = []
dy1 = []
dx2 = []
dy2 = []
dx3 = []
dy3 = []

for i in range(len(Dx)-1):
    dx.append(float(Dx[i] - Dx[i+1]))
    dy.append(float(Dy[i] - Dy[i+1]))

#Starting with the first value, the first is subtracted from the next value,
#giving us the difference in position in one axis. This is repeated for both axies

for i in range(int(len(Dx)/2)):
    dx1.append(float(Dx[2*i] - Dx[2*i+2]))
    dy1.append(float(Dy[2*i] - Dy[2*i+2]))

#Starting with the first value, the first is subtracted from the third value,
#giving us the difference in position in one axis. This is repeated for both axies

for i in range(int(len(Dx)/3)):
    dx2.append(float(Dx[3*i] - Dx[3*i+3]))
    dy2.append(float(Dy[3*i] - Dy[3*i+3]))

#Starting with the first value, the first is subtracted from the fourth value,
#giving us the difference in position in one axis. This is repeated for both axies

for i in range(int(len(Dx)/4)):
    dx3.append(float(Dx[4*i] - Dx[4*i+4]))
    dy3.append(float(Dy[4*i] - Dy[4*i+4]))

#Starting with the first value, the first is subtracted from the fifth value,
#giving us the difference in position in one axis. This is repeated for both axies

#Both pixel ratios are calculated. They are close but I'd like to keep them separate to keep it as accurate as possible

pixel_rx = (383*10**(-6))/640
pixel_ry = (283*10**(-6))/480

#The lists are then turned into arrays so that math can be done on them as a whole instead of using a for loop to do it.
    
delta_x = -1*np.array(dx)*pixel_rx
delta_y = -1*np.array(dy)*pixel_ry
delta_x1 = -1*np.array(dx1)*pixel_rx
delta_y1 = -1*np.array(dy1)*pixel_ry
delta_x2 = -1*np.array(dx2)*pixel_rx
delta_y2 = -1*np.array(dy2)*pixel_ry
delta_x3 = -1*np.array(dx3)*pixel_rx
delta_y3 = -1*np.array(dy3)*pixel_ry
```

```python
#plot of tragectory used
plt.plot(delta_x, delta_y)
plt.xlabel("Delta x (m)")
plt.ylabel("Delta y (m)")
plt.show()
```

```python
plt.hist(delta_x, 20)
plt.xlabel("Delta x (m)")
plt.ylabel("Number")
plt.show()
```

```python
plt.hist(delta_y, 20)
plt.xlabel("Delta y (m)")
plt.ylabel("Number")
plt.show()
```

The means and standard deviations were found for the two different histograms. They are both consistant with an overal motion of zero. 

```python
print("The mean of delta x is: " + str(np.mean(delta_x)))
print("The mean of delta y is: " + str(np.mean(delta_y)))
print("The standard deviation of x is: " + str(np.std(delta_x)))
print("The standard deviation of y is: " + str(np.std(delta_y)))
```

```python
#--------------3.6.3 Boltzmann Constant--------------------
```

In 3.3.2, equations for D given your standard deviation and time interval are shown. Rearranging those equations gives you
$D = \frac{\Delta r^{2}}{4\Delta t}$. Since our time step is 5 seconds the bottom becomes 20 and the top are the means of the two different axies sumed in quadrature to give us the average displacement squared on top. The uncertainty on D only relies on the uncertainty, or standard deviation, in our x an y measurements that give us $\Delta r^{2}$. Using error propigation we find the the uncertainty on $\Delta r^{2}$ is equal to $\sigma_{\Delta r^{2}} = \Delta r^{2} \sqrt{ \left (2\frac{\sigma_x}{x}\right )^{2} + \left(2\frac{\sigma_y}{y} \right )^{2}}$, and the uncertainty on D is equal to $\sigma_D = D \sqrt{\left(\frac{\sigma_{\Delta r^{2}}}{\Delta r^{2}} \right )^{2} }$. My x values are the largest factor in my uncertainty for D, followed by the y values. This makes sense because the positions of the particles has the largest effect on D regularly, being squared and added together.

Using another equation in 3.3.2 we can see that the Boltzmann constant is $k_b = \frac{Df}{T}$ and the uncertainty on $k_b$ is given by $\sigma_{k_b} = k_b \sqrt{\left( \frac{\sigma_D}{D} \right )^{2} + \left (\frac{\sigma_T}{T} \right )^{2} + \left (\frac{\sigma_R}{R}\right )^{2}}$. The systematic uncertainties that could be affecting $k_b$ could be movement of the microscope or the slide while the data was being taken, the actual movement of the particles--whether they were moving ore not, and also the resolution at which the data could be captured at. Trackpy is also not perfect and could skew the data slightly, giving inaccurate results. Most of these errors can be removed with better equipment. More data would also be useful as well.

```python
D1 = (np.mean(delta_x)**2 + np.mean(delta_y)**2)/20
delta_R2 = (np.mean(delta_x)**2 + np.mean(delta_y)**2)*np.sqrt((2*(np.std(delta_x))/np.mean(delta_x))**2 + (2*(np.std(delta_y))/np.mean(delta_y))**2 )
delta_D1 = D1*np.sqrt(((2*(np.std(delta_x))/np.mean(delta_x))**2) + (2*(np.std(delta_y))/np.mean(delta_y))**2)
delta_T = 0.2
delta_R = .2*10**(-6)
R = 2.7*10**(-6)
f = 6*np.pi*(1*10**(-3))*R
T = 296.75
kb = (D1*f)/T
delta_kb = kb*np.sqrt(((delta_D1/D1)**2) + ((delta_T/T)**2) + ((delta_R/R)**2))
# Df = kbT
print("The calculated value for the diffusion constant D is: " + str(D1) + " +/- " + str(delta_D1))
print("The calculated value for the Boltzmann constant is: " + str(kb) + " +/- " + str(delta_kb))
```

The value for the Boltzmann constant is 6 orders of magnitude off--yikes-- and about 3250 sigma away from the accepted value--double yikes. This is nowhere near the accepted value and I'm certain that this is because the data we collected was horrible. I would've liked to collect more data but was unable due to time constraints. 

```python
#---------------3.6.4 Time Evolution------------------------
```

```python
def rms(array):
    return np.sqrt((1/len(array))*(np.sum(array**2)))
    
```

```python
rms1 = []
rms2 = []
rms3 = []
rms4 = []
time1 = []
time2 = []
time3 = []
time4 = []
for i in range(1, len(delta_x)+1):
    rms1.append((10**8)*rms(delta_x[0:i]))
    time1.append(int(i - 1))
    
for i in range(1, len(delta_x1)+1):
    rms2.append((10**8)*rms(delta_x1[0:i]))
    time2.append(int(i*2 - 2))
    
for i in range(1, len(delta_x2)+1):
    rms3.append((10**8)*rms(delta_x2[0:i]))
    time3.append(int(i*3 - 3))
    
for i in range(1, len(delta_x3)+1):
    rms4.append((10**8)*rms(delta_x3[0:i]))
    time4.append(int(i*4 - 4))
    
plt.scatter(time1, rms1)
plt.scatter(time2, rms2)
plt.scatter(time3, rms3)
plt.scatter(time4, rms4)
plt.scatter(time3, .4*np.sqrt(time3))
plt.xlabel("Time (s)")
plt.ylabel("RMS x10^-8")
plt.legend(("Delta t = 5 sec", "Delta t = 10 sec", "Delta t = 15 sec", "Delta t = 20 sec", "Sqrt Delta t"))
plt.show()
```

The graph below is obviously wrong. There sould be a positive linear relationship that emerges from the data instead there is in negative linear relationship. The equation that gives this relationship is $\sigma^{2} = 2D\Delta t$, rearanging the equation yields $\pm \sigma = \sqrt{2D\Delta t}$. As the equation shows, $\sigma$ can have a negative relationship which in this case will be used just so we get a positive value for D.

```python
time_int = [10**.5,20**.5,30**.5,40**.5]

D2 = (np.mean(delta_x1)**2 + np.mean(delta_y1)**2)/40
D3 = (np.mean(delta_x2)**2 + np.mean(delta_y2)**2)/60
D4 = (np.mean(delta_x3)**2 + np.mean(delta_y3)**2)/80
delta_D2 = D1*np.sqrt(((2*(np.std(delta_x1))/np.mean(delta_x1))**2) + (2*(np.std(delta_y1))/np.mean(delta_y1))**2)
delta_D3 = D1*np.sqrt(((2*(np.std(delta_x2))/np.mean(delta_x2))**2) + (2*(np.std(delta_y2))/np.mean(delta_y2))**2)
delta_D4 = D1*np.sqrt(((2*(np.std(delta_x3))/np.mean(delta_x3))**2) + (2*(np.std(delta_y3))/np.mean(delta_y3))**2)

var = [delta_D1, delta_D2, delta_D3, delta_D4]

plt.scatter(time_int, var)
plt.xlabel("sqrt(Delta t)")
plt.ylabel("sigma")
coefficients = np.polyfit(time_int, var, 1)
y2 = np.polyval(coefficients, time_int)
print(coefficients)
plt.plot(time_int, y2)
plt.show()
```

The value for D found above is two orders of magnitude off from the value of D that was calculated further up. It is within one sigma of the calculated value however. The linear relationship is still objectively bad though.
